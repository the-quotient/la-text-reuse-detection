```python
import os
import json
import glob

def load_jsonl(path):
    with open(path, "r", encoding="utf-8") as f:
        return [json.loads(line)["sentence"] for line in f]

def count_stats(sentences):
    sent_count = len(sentences)
    word_count = sum(len(sentence.split()) for sentence in sentences)
    return sent_count, word_count

def collect_benchmark_data(query_path, candidate_folder):
    queries = load_jsonl(query_path)

    candidate_files = glob.glob(os.path.join(candidate_folder, "*.jsonl"))
    candidates = []
    for path in candidate_files:
        candidates.extend(load_jsonl(path))

    # Compute stats
    q_sent, q_word = count_stats(queries)
    c_sent, c_word = count_stats(candidates)

    pair_count = q_sent * c_sent

    stats = {
        "queries": {
            "sentences": q_sent,
            "words": q_word
        },
        "candidates": {
            "sentences": c_sent,
            "words": c_word
        },
        "total_pairs": pair_count
    }

    return stats

```
